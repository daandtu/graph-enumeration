```python
from time import time
import matplotlib.pyplot as plt
from graph_enumeration import GraphGenerator
```

```python
data = {}
for n_types in range(1, 10):
    node_types = list(range(n_types))
    for n_nodes in range(1, 7):
        print(f"Types: {n_types}, Nodes: {n_nodes}", end="\r", flush=True)
        enumerator = GraphGenerator(
            nodes=node_types, max_nodes=n_nodes, file_cache_path='../cache'
        )
        count = 0
        start_time = time()
        for graph in enumerator.iterator():
            count += 1
        data[(n_types, n_nodes)] = (count, time() - start_time)
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(11, 4), dpi=130)
for n_types in range(1, max([key[0] for key in data.keys()]) + 1):
    count_data = [(key[1], data[key][0]) for key in data.keys() if key[0] == n_types]
    time_data = [(key[1], data[key][1]) for key in data.keys() if key[0] == n_types]
    axs[0].scatter(*zip(*count_data), label=n_types)
    axs[1].scatter(*zip(*time_data), label=n_types)
axs[0].set(yscale='log', xlabel='# nodes', ylabel='# graphs')
axs[0].set_title('Number of graphs per number\nof nodes and node colors')
axs[1].set(yscale='log', xlabel='# nodes', ylabel='Time / s')
axs[1].set_title('Computation time per number\nof nodes and node colors')
axs[0].legend(title='# colors', ncols=2)
plt.savefig('graph_enumeration_performance.png', bbox_inches='tight')
plt.show()
```
